### Bubbles data

```python
years = list(range(2019,2023))
bubbles = []
for cost_label, values in expense_data["costs"].items():
    for name, amount_array in values.items():
        if name!="_factor":
            d = [{"cost_label":cost_label
                  , "year_total":
                  , "year":years[i]
                  , "name":name
                  , "amount":v} for i,v in enumerate(amount_array) ]
            bubbles.extend(d)
```

```python
with open ("data/bubbles.json","w") as f:
    json.dump(bubbles,f)
```
